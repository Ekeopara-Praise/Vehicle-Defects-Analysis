* **Select Some Specific Vechicle Defects**

The idea here is that we want to select specifically, those vechicle defects involving Toilets, Door and Window defects and Water defects. To achieve this, we basically classified all toilet related defects as 'Toilet', those related to Door & Window and Water as 'Sleeping room' and 'Water' respectively. 

```python
defects_df = vechicle_defects.copy()

defects_df.head(1)
```

```python
import re

# Function to map entries based on conditions
def map_entries(entry):
    if re.search(r'\b(?:Toilet|toilets|toilets|water|Freshwater|waters)\b', entry, flags=re.IGNORECASE):
        return 'Toilet'

    elif re.search(r'\b(?:sleeping rooms|door|doors|windows)\b', entry, flags=re.IGNORECASE):
        return 'Sleeping room'
        
    else:
        return entry

# Apply the function to the 'details' column
defects_df['details'] = defects_df['details'].apply(map_entries)
```

At this stage, the following machine learning model will be investigaated in the building of a machine learnig model for predictive maintenance; 

 **1.1 Random Forest**

The machine learning problem being solved involves the the prediction of a categorical output, and hence is a ``classification problem``.  **1.2 Description of model**

A random forest classifier is a type of machine learning classification algorithm which serves as a meta-estimator, employing multiple decision tree classifiers on different subsets of the dataset. The approach involves averaging the predictions to enhance predictive accuracy and mitigate overfitting. The size of the subsets is governed by the max_samples parameter when bootstrap is set to True (the default setting); otherwise, the entire dataset is utilized to construct each tree (Pedregosa, 2011).

 **1.3 Justification of model selection**

Random forest classifier's popularity is driven by its user-friendly nature and versatility, as it effectively manages tasks in both classification and regression scenarios (E R, 2021). The following reasons justifies the selection of Random Forest Classifier as a better model for this task;
- It is applicable for classification tasks, and that precisely aligns with the scenario we are addressing in this analysis.
- The model demonstrates robustness and effectively addresses the issue of overfitting by relying on either majority voting or averaging for generating the output.
- Its stability is notably high, given that it aggregates the average responses from a substantial number of trees.
- The model is resistant to the curse of dimensionality. Due to the fact that each tree does not take into account all attributes, there is a reduction in the feature space.
- The algorithm has large number of tunable hyperparameters which can be optimized for improved accuracy performance

## Model Evaluation Matrics

**Recall:** Recall is a metric that assesses the number of correct positive predictions out of all potential positive predictions. Unlike precision, which focuses solely on accurate positive predictions among all positive predictions, recall offers insight into missed positive predictions. $Recall$ = $TP/(TP+FN)$, where TP = True Positive, FN = False Negative (Jason Brownlee, 2020).

**Precision:** Precision is a metric that measures the accuracy of positive predictions. It is computed by dividing the number of correctly predicted positive examples by the total number of positive examples predicted. Usually ranges between 0 (no precision) to 1 (perfect score). $Precision$ = $TP/(TP+FP)$, where TP = True Positive, FP = False Positive (Jason Brownlee, 2020).

**F1 Score:** The F-Measure serves as a consolidated metric, blending precision and recall into a singular measure that encapsulates both characteristics. Also known as the F-Score or the F1-Score, it represents the harmonic mean of precision and recall. This metric is widely employed, particularly in addressing challenges posed by imbalanced classification problems (Jason Brownlee, 2020).

**Accuracy:** This metric computes the total correct predictions out of all the number of predictions made by the model. However, it won't be a priority metric as our data is imbalanced.

* **Handling Model Imbalance**

For handling the data imbalance, we will be using the SMOTE method. This technique, known as the Synthetic Minority Oversampling Technique (SMOTE), constitutes a form of data augmentation specifically tailored for the minority class. Is a straightforward strategy is to replicate instances within the minority class, but this doesn't introduce novel information to the model. Alternatively, a more sophisticated approach involves generating new instances by synthesizing them from the existing examples (Brownlee, 2020).

It can then implemented below;

```python
from imblearn.over_sampling import SMOTE

# Count the class distribution before SMOTE
print("Class distribution before SMOTE:", y.value_counts())

# Applying SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Count the class distribution after SMOTE
print("Class distribution after SMOTE:", y_resampled.value_counts())
```

```python
from sklearn.metrics import classification_report

# Generate the classification report as a dictionary
report_dict = classification_report(y_test, y_pred_rf, output_dict=True)

# Initialize lists to store metrics for class 0 and class 1
precision_list = []
recall_list = []
f1_score_list = []

# Specify the number of decimal places
decimal_places = 3

# Extract metrics for each class
for class_name, metrics in report_dict.items():
    if class_name.isdigit():  # Check if the class name is numeric (class labels)
        # Append precision, recall, and F1-score to their respective lists
        precision_list.append(round(metrics['precision'], decimal_places))
        recall_list.append(round(metrics['recall'], decimal_places))
        f1_score_list.append(round(metrics['f1-score'], decimal_places))

# --------------------------------------------------------------------------------------------------------------------#

# Set the figure size
fig = plt.subplots(figsize=(15, 8))

# Define parameters for the bar chart
N = 2
ind = np.arange(N)
width = 0.25

# Define data for the first set of bars (Precision scores)
index = ['Precision', 'Recall', 'F1']
Precision_score = precision_list
bar1T = plt.bar(ind, Precision_score, width, color='#6495ED')
for bar in bar1T:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom', ha='center', size='large')

# Define data for the second set of bars (Recall scores)
index = ['Precision', 'Recall', 'F1']
Recall_score = recall_list
bar2T = plt.bar(ind+width, Recall_score, width, color='#FFBF00')
for bar in bar2T:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom', ha='center', size='large')

# Define data for the third set of bars (F1 scores)
F1_score = f1_score_list
bar3T = plt.bar(ind+width*2, F1_score, width, color='#40E0D0')
for bar in bar3T:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom', ha='center', size='large')

# Set the y-axis limit
plt.ylim([0, 1.0])

# Set labels and title
plt.ylabel('Scores', fontsize=20)
plt.title("Performance Scores for Target Classes - After Handling Data Imbalance", fontsize=20)

# Set x-axis ticks and labels
plt.xticks(ind+width, ['Sleeping room', 'Toilet'], rotation=25, fontsize=20)

# Add legend
plt.legend(('Precision', 'Recall', 'F1'), loc='upper left', fontsize=13)

# Set plot style
sns.set_style("darkgrid")

# Display the plot
plt.show()
```
